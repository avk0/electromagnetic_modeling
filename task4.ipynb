Демонстрация численного метода вычисления уравнений Лапласа и Пуассона методом Гаусса–Зейделя в одномерном случае 

```python
import matplotlib.pyplot as plt
import numpy as np


x, h = np.linspace(0, 1, 101, retstep=True)

def ro(x):
    # заданная плотность заряда
    return 12 * x ** 2

def func(x):
    # аналитическое распределение потенциала
    return x - x ** 4

Number_of_Iteration = int(input("Количество итераций: "))
Omega = float(input("Omega: "))
phi = np.zeros(x.size) 
Esym = np.zeros(Number_of_Iteration)

for j in range(Number_of_Iteration):
    E = 0

    for i in range(x.size - 2):
        phi[i + 1] = ((1 - Omega) * phi[i + 1] + 0.5 * Omega * (phi[i + 2] + phi[i] + h ** 2
        * ro(x[i + 1])))

    if j % 50 == 0:
        plt.plot(x, phi, '--', linewidth=1.2)

    for i in range(x.size - 1):
        E += (phi[i + 1] - phi[i]) ** 2 / (2 * h) - h * ro(x[i]) * phi[i]  # вычисление энергии
    Esym[j] = E
    
plt.plot(x, func(x))
plt.show()

# Энергия
A3 = -9/14 * np.ones(Number_of_Iteration)
plt.plot(Esym)
plt.plot(A3)
plt.show()
```
